```python
import jieba
from collections import Counter
from pyecharts.charts import WordCloud

jieba.add_word('蔡徐坤')

# 直接定义停用词列表，不需要读取文件了
swords = ['的', '了', '在', '是', '我', '有', '和', '就', 
          '不', '人', '都', '一', '一个', '上', '也', '很', 
          '到', '说', '要', '去', '你', '会', '着', '没有', 
          '看', '好', '自己', '这', '转发微博', '回复']
```

```python
from collections import Counter


def plot_word_cloud(data, swords, columns):
    # 1. 数据清洗：转字符串并处理空值
    text = ''.join(data[columns].fillna('').astype(str))
    
    # 2. 分词
    words = list(jieba.cut(text))
    
    # 3. 去除停用词
    ex_sw_words = []
    for word in words:
        if len(word) > 1 and (word not in swords):
            ex_sw_words.append(word)
    
    # 4. 统计词频
    c = Counter(ex_sw_words)
    wc_data = pd.DataFrame({'word': list(c.keys()), 'counts': list(c.values())})
    wc_data = wc_data.sort_values(by='counts', ascending=False).head(100)
    
    # 修正1：准备数据格式。新版 add 需要传入 [(词, 数), (词, 数)...] 的列表
    data_pair = [list(z) for z in zip(wc_data['word'], wc_data['counts'])]
    
    # 修正2：设置画布大小。新版使用 init_opts=opts.InitOpts(...)
    wordcloud = WordCloud(init_opts=opts.InitOpts(width="1300px", height="620px"))
    
    # 修正3：添加数据。参数名变成了 data_pair
    wordcloud.add(series_name="", data_pair=data_pair, word_size_range=[20, 100])
    
    # 设置标题（可选）
    wordcloud.set_global_opts(title_opts=opts.TitleOpts(title="词云分析"))
    
    # --- 关键修改点结束 ---
    
    return wordcloud
```

```python
wc=plot_word_cloud(data=data_true, swords=swords, columns='user_description')
wc.render('wordcloud.html')
```
